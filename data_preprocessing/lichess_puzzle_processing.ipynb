```python
import pandas as pd

puzzle_df = pd.read_csv('lichess_db_puzzle.csv.zst', compression='zstd')
puzzle_df['FEN'] = puzzle_df['FEN'].apply(lambda fen: ' '.join(fen.split(' ')[:4]))
puzzle_df['OpeningTags'] = puzzle_df['OpeningTags'].fillna('')
```

```python
unique_tags = set()
for tags in puzzle_df['OpeningTags'].unique():
    unique_tags |= set(tags.split(' '))
unique_tags -= {''}
```

```python
for tag in unique_tags:
    if 'kings' in tag.lower():
        print(tag)    
```

```python
for tag in unique_tags:
    for name in ['dutch', 'english', 'sicilian', 'french', 'benoni', 'queens_gambit', 'kings_indian']:
        tag_components = tag.lower().split('_')
        name_components = name.split('_')
        if len(name_components) > len(tag_components): 
            continue
        tag_components = tag_components[: len(name_components)]
        if '_'.join(name_components) == '_'.join(tag_components):
            print(tag)
```

```python
def detect_opening(row, opening_name):
    tags = row['OpeningTags'].split(' ')
    
    for tag in tags:
        tag_components = tag.lower().split('_')
        name_components = opening_name.split('_')
        if len(name_components) > len(tag_components): 
            continue
        tag_components = tag_components[: len(name_components)]
        if '_'.join(name_components) == '_'.join(tag_components):
            return 1
        
    return 0
    
openings_df = puzzle_df[['PuzzleId', 'FEN', 'OpeningTags']].copy()

for opening_name in ['dutch', 'english', 'sicilian', 'french', 'benoni', 'queens_gambit', 'kings_indian']:
    print(f'Finding {opening_name}')
    openings_df[opening_name] = openings_df.apply(detect_opening, args=(opening_name,), axis=1)
```

```python
openings_df_refined = openings_df[openings_df.iloc[:, 3:].sum(axis = 1) >= 1]
```

```python
openings_df_refined.to_csv('lichess_puzzles_openings.csv')
```
